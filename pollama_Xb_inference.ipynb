```python
from peft import PeftModel, PeftConfig
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from pprint import pprint

config = PeftConfig.from_pretrained(f"szymonrucinski/krakowiak-{MODEL_VERSION}")
model = AutoModelForCausalLM.from_pretrained(f"meta-llama/Llama-2-{MODEL_VERSION}-hf", use_auth_token="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",load_in_4bit=True)

tokenizer = AutoTokenizer.from_pretrained(f"meta-llama/Llama-2-{MODEL_VERSION}-hf", use_auth_token="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")
model = PeftModel.from_pretrained(model, f"szymonrucinski/krakowiak-{MODEL_VERSION}")

pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    device_map="auto",
)
```

```python
import gradio as gr

def greet(query):
    sequences = pipeline(
    f'### Użytkownik: {query} ### Asystent: ',
    # do_sample=True,
    # top_k=10,
    top_p=0.95,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=200,
    temperature=0.2,
    repetition_penalty=2.0,
    # length_penalty=-0.5
  )
    for seq in sequences:
      return f"{seq['generated_text']}".split("Asystent:")[1]

demo = gr.Interface(fn=greet, inputs="text", outputs="text")

demo.launch()
```
